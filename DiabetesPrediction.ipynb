```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
df
```

```python
for i in range(25):
    nullCheck = df.iloc[:,i].isnull().sum()
    print("The number of null value in column ",i," are: ",nullCheck)
```

```python
Y = df['Outcome']
Y
```

```python
X = df.drop(['Outcome'],axis='columns')
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=7)
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,Y_train)
Y_predict = lr.predict(X_test)

Y_predict
```

```python
Y_test
```

```python
print("The Predicted value of Y is :")
print(Y_predict[:10])
print("The test value of Y is :")
print(np.array(Y_test[:10]))
print("The accuracy of Logistic Regression is :")
print(lr.score(X_test,Y_test))
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn import metrics
matrix = metrics.confusion_matrix(Y_predict,Y_test)
sns.heatmap(matrix,annot=True,cmap='Blues',fmt='0.1f')
plt.xlabel("Predicted Values")
plt.ylabel("Test Value")
plt.title("Confusion Matrix for Logistic Regression")
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, Y_train)
Y_predict = nb.predict(X_test)
print("predicted value of Y")
print(Y_predict[:10])
print("test values of Y ")
print(np.array(Y_test[:10]))
print("The Accuracy of the Naive Bayes is ")
print(nb.score(X_test,Y_test))
```

```python
matrix = metrics.confusion_matrix(Y_predict,Y_test)
sns.heatmap(matrix,annot=True,cmap='Blues',fmt='0.1f')
plt.xlabel("Predicted Values")
plt.ylabel("Test Value")
plt.title("Confusion Matrix for Naive Bayes")
```

```python
from sklearn.model_selection import cross_val_score
lr.fit(X,Y)
print(cross_val_score(lr,X,Y,cv=10))
print(np.average(cross_val_score(lr,X,Y,cv=10)))
```

```python
nb.fit(X,Y)
print(cross_val_score(nb,X,Y,cv=10))
print(np.average(cross_val_score(nb,X,Y,cv=10)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
Y_predict = knn.predict(X_test)
Y_predict
```

```python
matrix = metrics.confusion_matrix(Y_predict,Y_test)
sns.heatmap(matrix,annot=True,cmap='Blues',fmt='0.1f')
plt.xlabel("Predicted Values")
plt.ylabel("Test Value")
plt.title("Confusion Matrix for KNeighborsClassifier")
```
